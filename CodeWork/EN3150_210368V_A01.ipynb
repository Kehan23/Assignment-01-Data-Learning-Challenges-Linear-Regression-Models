```python
import numpy as np
import matplotlib . pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Generate 100 samples
n_samples = 100

# Generate X values ( uniformly distributed between 0 and 10)
X = 10*np.random.rand(n_samples , 1)

# Generate epsilon values ( normally distributed with mean 0 and standard deviation 15)
epsilon = np.random.normal(0 , 15 , n_samples )

# Generate Y values using the model Y = 3 + 3X + epsilon
Y = 3 + 2*X + epsilon[: , np.newaxis]

```

2.

```python
r = np.random.randint (104)

# Split the data into training and test sets (80% train ,20% test )
X_train , X_test , Y_train , Y_test = train_test_split (X , Y ,test_size =0.2 , random_state = r )

# Plot the data points
plt.figure(figsize =(10 , 6))
plt.scatter(X_train , Y_train , alpha =1 , marker ='o', color ='red', label ='Training Data')
plt.scatter(X_test , Y_test , alpha =1 , marker ='s', color ='blue', label ='Testing Data')
plt.show()
```

```python
d = X_train.shape[1]  # Number of independent features in the training data (i.e., the number of columns)

n = len(Y_train)  # Number of observations (i.e., the number of rows) in the training data

# Calculate predictions for the testing data
y_hat = model.predict(X_test)  # Predicted values for the test set using the trained model

Residuals = Y_test.to_numpy() - y_hat  # Calculate residuals (actual test values - predicted values)

# Residual Sum of Squares (RSS)
RSS = np.sum(Residuals ** 2)  # Sum of squared residuals, measures the total deviation of the predictions from the actual values

# Residual Standard Error (RSE)
RSE = np.sqrt((1 / (n - d - 1)) * RSS)  # Estimate of the standard deviation of the residuals, adjusted for the number of predictors

# Mean Squared Error (MSE)
MSE = RSS / n  # Average squared difference between actual and predicted values; measures the model's prediction accuracy

# Total Sum of Squares (TSS)
TSS = np.sum((Y_train.to_numpy() - np.mean(Y_train.to_numpy()))**2)  # Total variance in the training data, a measure of how spread out the actual values are

# R-squared statistic (R²)
R_Squared = (TSS - RSS) / TSS  # Proportion of the variance in the dependent variable explained by the model; indicates the model's goodness of fit

# Print results
print(f"Residual Sum of Squares (RSS): {RSS}")  # Output the RSS value, indicating total error
print(f"Residual Standard Error (RSE): {RSE}")  # Output the RSE value, providing an estimate of the average size of the residuals
print(f"Mean Squared Error (MSE): {MSE}")  # Output the MSE value, reflecting the model's average prediction error
print(f"R_Squared: {R_Squared}")  # Output the R² value, showing the percentage of variance explained by the model
```

```python
import pandas as pd
import statsmodels.api as sm

# Ensure to use the training set for fitting the model
intercept = np.ones((X_test.shape[0], 1))  # Create a column of ones for the intercept
X_test_with_intercept = np.hstack((intercept, X_test))  # Add intercept to X_test



# Build the OLS (Ordinary Least Squares) regression model
ols_regression = sm.OLS(Y_test, X_test_with_intercept).fit()

# Generate the model's summary statistics
model_summary = ols_regression.summary()

# Extract the p-values for each feature from the model summary
# Note: model_summary.tables[1] is a table, so you should not directly use it to extract p-values.
# Use ols_regression.pvalues instead.
p_values_df = pd.DataFrame({
    'Feature': ['Intercept'] + list(X_test.columns),  # Adjust this if X_test is a DataFrame
    'P-Value': ols_regression.pvalues
})

# Display the summary and the DataFrame
print(model_summary)
print(p_values_df)
```
